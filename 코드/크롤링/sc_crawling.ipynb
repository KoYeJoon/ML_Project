```python
import re
cmmt_list=[]

for b in cmmt_box :
    comment=b.find('div',attrs={'class':'commentItem__body sc-hyphenate'})
    comment=comment.getText().split('\n')[0]
    if comment=='' :continue
    
    id_name=b.find('a',attrs={'class':'commentItem__usernameLink sc-link-light'})
    id_name=id_name.getText().strip()
    if id_name=="":continue
        
    date=b.find('a',attrs={'class':'commentItem__timestampLink sc-link-light'})
    date=date.getText().split('\n')[0]
    if date=='':continue
        
    image=b.find(attrs={'aria-role':'img'})
    image=image.get('style')
    try:
        found = re.search(r"", image).group(1)
    except AttributeError:
        found = '' 
        
    cmmt={}
    cmmt['id']=id_name
    cmmt['date']=date
    cmmt['image_src']=found
    cmmt['content']=comment
    
    cmmt_list.append(cmmt)

print(cmmt_list)
```
