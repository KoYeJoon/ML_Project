```python
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time 

#Ïù¥ÎØ∏ÏßÄ ÌÅ¨Î°§ÎßÅ


body = driver.find_element_by_tag_name('body')

#Ïù∏Í∏∞Ïàú/ÏûëÏÑ±Ïàú ÏÑ†ÌÉùÌï† Ïàò ÏûàÎäî ÏòÅÏó≠ ÌÅ¥Î¶≠
#driver.find_element_by_xpath('//paper-button[@class="dropdown-trigger style-scope yt-dropdown-menu"]').click()
#Ïù∏Í∏∞Ïàú Ïπ¥ÌÖåÍ≥†Î¶¨ ÌÅ¥Î¶≠
#driver.find_element_by_xpath('//paper-listbox[@class="dropdown-content style-scope yt-dropdown-menu"]/a[1]').click()

num_pagedown =100 #nÎ≤à Î∞ëÏúºÎ°ú ÎÇ¥Î¶¨Îäî Í≤É
while num_pagedown:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    num_pagedown -=1
    
page=driver.page_source
soup=BeautifulSoup(page,'html.parser')

#comments=soup.find_all('yt-formatted-string',attrs={'class':'style-scope ytd-comment-renderer'})

cmmt_box=soup.find_all('ytd-comment-renderer',attrs={'id':'comment'})
real=soup.find('video')
real=real.get('src')

print(real)
```

```python

cmmt_list=[]

for b in cmmt_box :
    comment=b.find('yt-formatted-string',attrs={'id':'content-text'})
    comment=comment.getText().split('\n')[0]
    if comment=='' :continue
    
    id_name=b.find('a',attrs={'id':'author-text'})
    id_name=id_name.getText().strip()
    if id_name=="":continue
        
    date=b.find('a',attrs={'class':'yt-simple-endpoint style-scope yt-formatted-string'})
    date=date.getText().split('\n')[0]
    if date=='':continue
        
    image=b.find('img')
    image=image.get('src')
    
    if image=='':continue
        
    cmmt={}
    cmmt['id']=id_name
    cmmt['date']=date
    cmmt['image_src']=image
    cmmt['content']=comment
    
    cmmt_list.append(cmmt)

print(cmmt_list)
```
